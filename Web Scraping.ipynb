## Scrapear listado de productos

```python
from selenium import webdriver
from bs4 import BeautifulSoup

webdriver = webdriver.Chrome()

def get_product_list():
    #Usar url de los productos que se quieran obtener
    url = "https://www.amazon.com.mx/s?k=nintendo&crid=1MMYIZHOLTOG9&sprefix=nintn%2Caps%2C140&ref=nb_sb_ss_ts-doa-p_2_5"
    webdriver.get(url)
    webdriver.refresh()

    soup = BeautifulSoup(webdriver.page_source, "lxml")
    #click a la clase, copiar y pegar
    a_class = "a-link-normal s-line-clamp-4 s-link-style a-text-normal"
    a_tags = soup.find_all("a", class_ = a_class)

    for a in a_tags:
        print("titulo:", a.text.replace("\n", ""))
        print("link:", a.get("href"))
        print("\n")

get_product_list()

webdriver.close()
```

## Scrapear detalle de productos

```python
from selenium import webdriver
from bs4 import BeautifulSoup

webdriver = webdriver.Chrome()

def get_product_list():
    #Usar url de los productos que se quieran obtener
    url = "https://www.amazon.com.mx/s?k=nintendo&crid=1MMYIZHOLTOG9&sprefix=nintn%2Caps%2C140&ref=nb_sb_ss_ts-doa-p_2_5"
    webdriver.get(url)
    webdriver.refresh()

    soup = BeautifulSoup(webdriver.page_source, "lxml")
    #click a la clase, copiar y pegar
    a_class = "a-link-normal s-line-clamp-4 s-link-style a-text-normal"
    a_tags = soup.find_all("a", class_ = a_class)

    for a in a_tags:
        link = a.get("href")
        print("titulo:", a.text.replace("\n", ""))
        print("link:", link)
        
        detail_link = "https://www.amazon.com.mx/" + link if "https" not in link else link
        get_details(detail_link)
        print("\n")
    webdriver.close()

def get_details(url):
    webdriver.get(url)
    webdriver.refresh()
    soup = BeautifulSoup(webdriver.page_source, "lxml")
    price = float(soup.find("span", class_ = "a-price-whole").text.replace(",", ""))
    price += float(f"0.{soup.find("span", class_ = "a-price-fraction").text}")
    print(price)

    description = soup.find("div", id = "feature-bullets").text
    print(description)

get_product_list()
```
